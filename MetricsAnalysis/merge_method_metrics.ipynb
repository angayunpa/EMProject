```python
import pandas as pd
```

```python
df1 = pd.read_csv("/content/method_metricsSM.csv")
```

```python
df2 = pd.read_csv("/content/method_metrics_radon.csv").drop(columns=["Unnamed: 0"])
```

```python
method_metrics = pd.concat([df1, df2], axis=1)
method_metrics
```

```python
method_metrics.to_csv("method_metrics_all.csv", index=False)
```

```python
method_metrics.describe()
```

```python
method_metrics.describe().to_csv("method_metrics_all_describe.csv", index=False)
```

```python
from matplotlib import pyplot as plt
import seaborn as sns

plt.figure(figsize=(20, 20))
corr = method_metrics.corr(numeric_only=True)
sns.heatmap(corr, annot=True)
```

```python
method_metrics.corr(numeric_only=True).to_csv("method_metrics_all_corr.csv", index=False)
```
