```python
import pandas as pd
```

```python
df1 = pd.read_csv("../data/method_metricsSM.csv")
```

```python
df2 = pd.read_csv("../data/method_metrics_radon.csv").drop(columns=["Unnamed: 0"])
```

```python
method_metrics = df1.merge(df2, on='repository', how='inner')
method_metrics
```

```python
method_metrics[['repository']].to_csv('../data/repo_list.csv')
```

```python
method_metrics.to_csv("../data/method_metrics_all.csv", index=False)
```

```python
method_metrics.describe()
```

```python
method_metrics.describe().to_csv("../data/method_metrics_all_describe.csv", index=False)
```

```python
from matplotlib import pyplot as plt
import seaborn as sns

plt.figure(figsize=(20, 20))
corr = method_metrics.corr(numeric_only=True)
sns.heatmap(corr, annot=True)
```

```python
method_metrics.corr(numeric_only=True).to_csv("../data/method_metrics_all_corr.csv", index=False)
```
